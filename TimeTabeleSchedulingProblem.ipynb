```python
import copy
import random
```

### Data

```python
def get_Coures ():
    return copy.deepcopy(["Computer Graphics", "Cybersecurity Essentials", "Digital Signal Processing", "Game Development", "Human-Computer Interaction", "Internet of Things", 
    "Parallel and Distributed Computing", "Robotics Fundamentals", "Virtual Reality", "Web Development", "Bioinformatics", "Computer Vision",
    "Cloud Computing", "Embedded Systems", "Natural Language Processing", "Blockchain Technology", "Quantum Computing", "Cryptography",
    "Data Mining", "Network Security", "Information Retrieval", "Geographic Information Systems", "Wireless Communication", "Big Data Analytics"])

def get_Sections():
    return copy.deepcopy(['W', 'X', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'U', 'K'])

def get_Professors():
    return copy.deepcopy([
    "Ayesha Ahmed", "Bilal Khan", "Farah Malik", "Kamran Ali", "Nadia Iqbal", "Omar Shah",
    "Sadia Khan", "Tariq Mahmood", "Uzma Batool", "Waqas Ali", "Xavier Rehman", "Yasmeen Abbasi"])


def get_Timeslots():
    return copy.deepcopy(["8:30-9:50", "10:05-11:25", "11:40-1:00", "1:15-2:30", "2:45-4:05", "4:20-5:35"])

def get_Rooms():
    return copy.deepcopy(["D-301", "E-202", "F-305", "G-401", "H-102", "J-404"])
```

```python
Courses = get_Coures()
Sections = get_Sections()
Professors = get_Professors()
Timeslots = get_Timeslots()
Rooms = get_Rooms()
```

### Encoding

```python
def encode_Course_name(course_name):
    return copy.deepcopy([1 if course == course_name else 0 for course in Courses])

def enncode_lecture_type(lecture_type):
    return copy.deepcopy([1 if lecture_type == 'Theory' else 0])

def encode_section(section):
    return copy.deepcopy([1 if section_ == section else 0 for section_ in Sections])

def encode_section_strnegtn(section_strngth):
    return copy.deepcopy([int(bit) for bit in bin(section_strngth)[2:].zfill(7)])

def encode_professor(professor):
    return copy.deepcopy([1 if professor_ == professor else 0 for professor_ in Professors])

def encode_lecture_day(first_lecture_day):
    days_list =  ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
    return copy.deepcopy([1 if day == first_lecture_day else 0 for day in days_list])

def encode_lecture_timeslot(first_lecture_timeslot):
    return copy.deepcopy([1 if timeslot == first_lecture_timeslot else 0 for timeslot in Timeslots])

def encode_lecture_room(first_lecture_room):
    return copy.deepcopy([1 if room == first_lecture_room else 0 for room in Rooms])

def encode_lecture_room_size(first_lecture_room_size):
    return copy.deepcopy([1 if first_lecture_room_size == 60 else 0])

def encode_lecture_day(first_lecture_day):
    days_list =  ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
    return copy.deepcopy([1 if day == first_lecture_day else 0 for day in days_list])
    

def encode_chromosome(course_name, lecture_type, section, section_strength, professor, first_lecture_day, first_lecture_timeslot, first_lecture_room, first_lecture_room_size, second_lecture_day, second_lecture_timeslot, second_lecture_room, second_lecture_room_size):
    binary_encoded_chromosome = []
    binary_encoded_chromosome.extend(encode_Course_name(course_name))
    binary_encoded_chromosome.extend(enncode_lecture_type(lecture_type))
    binary_encoded_chromosome.extend(encode_section(section))
    binary_encoded_chromosome.extend(encode_section_strnegtn(section_strength))
    binary_encoded_chromosome.extend(encode_professor(professor))
    binary_encoded_chromosome.extend(encode_lecture_day(first_lecture_day))
    binary_encoded_chromosome.extend(encode_lecture_timeslot(first_lecture_timeslot))
    binary_encoded_chromosome.extend(encode_lecture_room(first_lecture_room))
    binary_encoded_chromosome.extend(encode_lecture_room_size(first_lecture_room_size))
    binary_encoded_chromosome.extend(encode_lecture_day(second_lecture_day))
    binary_encoded_chromosome.extend(encode_lecture_timeslot(second_lecture_timeslot))
    binary_encoded_chromosome.extend(encode_lecture_room(second_lecture_room))
    binary_encoded_chromosome.extend(encode_lecture_room_size(second_lecture_room_size))
    return binary_encoded_chromosome
```

### Decoding

```python
def decode_course(chromosome):
    return Courses[chromosome[0:0 + len(Courses)].index(1)]

def decode_course_type(chromosome):
    type_index = chromosome[len(Courses):len(Courses) + 1]
    if type_index[0] == 1:
        return 'Theory'
    else:
        return 'Lab'

def decode_section(chromosome):
    return Sections[chromosome[len(Courses) + 1:len(Courses) + 1 + len(Sections)].index(1)]

def decode_section_strength(chromosome):
    section_strength_index = chromosome[len(Courses) + 1 + len(Sections):len(Courses) + 1 + len(Sections) + 7]
    two = 2
    return int(''.join(map(str, section_strength_index)), two)

def decode_professor(chromosome):
    index = len(Courses) + 1 + len(Sections) + 7
    professor_index =  chromosome[len(Courses) + 1 + len(Sections) + 7:len(Courses) + 1 + len(Sections) + 7 + len(Professors)]
    try:
        ret_ind = professor_index.index(1)
        return Professors[ret_ind]
    except ValueError:
        r = random.randint(index, index + len(Professors) - 1)
        chromosome[r] = 1
        return Professors[r - index]

def decode_first_lecture_day(chromosome):
    start_index = len(Courses) + 1 + len(Sections) + 7 + len(Professors)
    end_index = start_index + 5
    first_lecture_day_section = chromosome[start_index:end_index]
    day_mapping = {0: "Monday", 1: "Tuesday", 2: "Wednesday", 3: "Thursday", 4: "Friday"}
    for index, value in enumerate(first_lecture_day_section):
        if value == 1:
            return day_mapping[index]

def decode_first_lecture_timeslot(chromosome):
    return Timeslots[chromosome[len(Courses) + 1 + len(Sections) + 7 + len(Professors) + 5:len(Courses) + 1 + len(Sections) + 7 + len(Professors) + 5 + len(Timeslots)].index(1)]

def decode_first_lecture_room(chromosome):
    return Rooms[chromosome[len(Courses) + 1 + len(Sections) + 7 + len(Professors) + 5 + len(Timeslots):len(Courses) + 1 + len(Sections) + 7 + len(Professors) + 5 + len(Timeslots) + len(Rooms)].index(1)]

def decode_first_lecture_room_size(chromosome):
    ind = chromosome[len(Courses) + 1 + len(Sections) + 7 + len(Professors) + 5 + len(Timeslots) + len(Rooms):len(Courses) + 1 + len(Sections) + 7 + len(Professors) + 5 + len(Timeslots) + len(Rooms) + 1]
    return 60 if ind[0] == 1 else 120

def decode_second_lecture_day(chromosome):
    ind = chromosome[len(Courses) + 1 + len(Sections) + 7 + len(Professors) + 5 + len(Timeslots) + len(Rooms) + 1:len(Courses) + 1 + len(Sections) + 7 + len(Professors) + 5 + len(Timeslots) + len(Rooms) + 1 + 5]
    return ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"][ind.index(1)]

def decode_second_lecture_timeslot(chromosome):
    return Timeslots[chromosome[len(Courses) + 1 + len(Sections) + 7 + len(Professors) + 5 + len(Timeslots) + len(Rooms) + 1 + 5:len(Courses) + 1 + len(Sections) + 7 + len(Professors) + 5 + len(Timeslots) + len(Rooms) + 1 + 5 + len(Timeslots)].index(1)]

def decode_second_lecture_room(chromosome):
    return Rooms[chromosome[len(Courses) + 1 + len(Sections) + 7 + len(Professors) + 5 + len(Timeslots) + len(Rooms) + 1 + 5 + len(Timeslots):len(Courses) + 1 + len(Sections) + 7 + len(Professors) + 5 + len(Timeslots) + len(Rooms) + 1 + 5 + len(Timeslots) + len(Rooms)].index(1)]

def decode_second_lecture_room_size(chromosome):
    ind = chromosome[len(Courses) + 1 + len(Sections) + 7 + len(Professors) + 5 + len(Timeslots) + len(Rooms) + 1 + 5 + len(Timeslots) + len(Rooms):len(Courses) + 1 + len(Sections) + 7 + len(Professors) + 5 + len(Timeslots) + len(Rooms) + 1 + 5 + len(Timeslots) + len(Rooms) + 1]
    return 60 if ind[0] == 1 else 120


def decode_chromosome(chromosome):

    decoded_chromosome = []
    temp = list()
    temp.append(decode_course(chromosome))
    temp.append(decode_course_type(chromosome))
    temp.append(decode_section(chromosome))
    temp.append(decode_section_strength(chromosome))
    temp.append(decode_professor(chromosome))
    temp.append(decode_first_lecture_day(chromosome))
    temp.append(decode_first_lecture_timeslot(chromosome))
    temp.append(decode_first_lecture_room(chromosome))
    temp.append(decode_first_lecture_room_size(chromosome))
    temp.append(decode_second_lecture_day(chromosome))
    temp.append(decode_second_lecture_timeslot(chromosome))
    temp.append(decode_second_lecture_room(chromosome))
    temp.append(decode_second_lecture_room_size(chromosome))

    for i in temp:
        decoded_chromosome.append(i)
    return decoded_chromosome
```

### Generating Random Chromosome

### Conflict Functions

```python
def same_room_during_same_timeslot(course_name, day, timeslot, room, room_usage):
    conflicts = 0
    if (day, timeslot, room) in room_usage:
        conflicts += 1
    else:
        triplet = (day,timeslot,room)
        room_usage[triplet] = course_name

    return conflicts

def sufficient_room_capacity(section_strength, capacity, room):
    conflicts = 0
    if section_strength > capacity:
        conflicts += 1

    return conflicts

def professor_course_count(professor, professor_courses):
    conflicts = 0
    if professor in professor_courses:
        professor_courses[professor] += 1
    else:
        professor_courses[professor] = 1
    
    if professor_courses[professor] > 3:
            conflicts += 1

    return conflicts

def section_course_count(course_name, section, section_courses):
    conflicts = 0
    if section in section_courses:
        section_courses[section].add(course_name)
    else:
        section_courses[section] = {course_name}

    if len(section_courses[section]) > 5:
        conflicts += 1

    return conflicts

def professor_assigned_to_two_lectures_same_time(day, timeslot, population, professor):
    conflicts = 0
    unique_conflicts = set()
    i = 0

    while i < len(population):
        decoded_c = decode_chromosome(population[i])
        if professor == decoded_c[4] and ((timeslot == decoded_c[6] and day == decoded_c[5]) or (timeslot == decoded_c[10] and day == decoded_c[9])):
            unique_conflicts.add(f"{professor} is assigned two lectures at the same time at {timeslot} on {day}")
        i += 1


    length_of_unique_conflicts = len(unique_conflicts)
    conflicts +=  length_of_unique_conflicts

    return conflicts

    
def same_section_assigned_different_rooms_at_same_time(day, timeslot, room, population, section):
    conflicts = 0
    unique_conflicts = set()
    i = 0
    while i < len(population):
        decoded_c = decode_chromosome(population[i])
        if section == decoded_c[2] and ((timeslot == decoded_c[6] and day == decoded_c[5] and room != decoded_c[7]) or
                (timeslot == decoded_c[10] and day == decoded_c[9] and room != decoded_c[11])):
            unique_conflicts.add(f"Section {section} is assigned different rooms at same time at {timeslot} on {day}")
        i += 1
 
    conflicts += len(unique_conflicts)
    
    return conflicts


def same_room_assigned_to_different_sections_at_same_time(day, timeslot, room, population,  section):
    conflicts = 0
    unique_conflicts = set()
    i = 0

    while i < len(population):
        decoded_c = decode_chromosome(population[i])
        if room == decoded_c[7] and ((timeslot == decoded_c[6] and day == decoded_c[5] and room == decoded_c[7]) or
                (timeslot == decoded_c[10] and day == decoded_c[9] and room == decoded_c[11]) and section != decoded_c[2]):
            unique_conflicts.add((section, timeslot, day, room))
        i += 1
        

    length_of_unique_ = len(unique_conflicts)
    conflicts += length_of_unique_

    return conflicts


def lab_lectures_two_consective_time_slots(lecture_type, timeslot1, timeslot2, day1, day2, room1, room2):
    conflicts = 0
    
    if day1 != day2:
        conflicts += 1
    if room1 != room2:
        conflicts += 1
    elif lecture_type.lower() == 'Lab' and timeslot2 is not None:
        timeslot1_endhour, timeslot1_endminute = map(int, timeslot1.split('-')[1].split(':'))
        timeslot2_starthour, timeslot2_startminute = map(int, timeslot2.split('-')[0].split(':'))

        timeslot1_minutes = timeslot1_endhour * 60 + timeslot1_endminute
        timeslot2_minutes = timeslot2_starthour * 60 + timeslot2_startminute

        if (timeslot1_minutes + 15 != timeslot2_minutes):
            conflicts += 1

    return conflicts        

def check_Course_Conflict(schedule_days, indexing):
    return abs(indexing[schedule_days[1]]) - indexing[schedule_days[0]] > 1

def course_have_two_lectures_per_week_not_on_same_or_adjacent_days(course, population):
    conflicts = 0
    course_schedule = set()
    i = 0

    while i < len(population):
        decoded_c = decode_chromosome(population[i])
        if decoded_c[0] == course:
            course_schedule.add((decoded_c[5]))
        i += 1

    indexing = {"Monday": 1, "Tuesday": 2, "Wednesday": 3, "Thursday": 4, "Friday": 5}

    max_lectures = 2
    if len(course_schedule) == max_lectures:
        schedule_days = [d for d in course_schedule] 
        if check_Course_Conflict(schedule_days, indexing):
            conflicts += 1
    else:
        conflicts += 1

    return conflicts
```

### Calculating Conflicts


```python
def calculate_conflicts(chromosome, population, room_usage, professor_courses, section_courses):
    conflicts = 0

    #Hard constraints Implementation
    
    #Classes can only be scheduled in free classroom.
    conflicts += same_room_during_same_timeslot(decode_course(chromosome), decode_first_lecture_day(chromosome), decode_first_lecture_timeslot(chromosome), decode_first_lecture_room(chromosome), room_usage)
    conflicts += same_room_during_same_timeslot(decode_course(chromosome), decode_first_lecture_day(chromosome), decode_first_lecture_timeslot(chromosome), decode_first_lecture_room(chromosome), room_usage)

    #classroom should be big enough to accommodate the section.
    conflicts += sufficient_room_capacity(decode_section_strength(chromosome), decode_first_lecture_room_size(chromosome), decode_first_lecture_room(chromosome))
    conflicts += sufficient_room_capacity(decode_section_strength(chromosome), decode_second_lecture_room_size(chromosome), decode_second_lecture_room(chromosome))

    #professor should not be assigned two different lectures at the same time.
    conflicts += professor_assigned_to_two_lectures_same_time(decode_first_lecture_day(chromosome), decode_first_lecture_timeslot(chromosome), population, decode_professor(chromosome))
    conflicts += professor_assigned_to_two_lectures_same_time(decode_second_lecture_day(chromosome), decode_second_lecture_timeslot(chromosome), population, decode_professor(chromosome))

    #The same section cannot be assigned to two different rooms at the same time
    conflicts += same_section_assigned_different_rooms_at_same_time(decode_first_lecture_day(chromosome), decode_first_lecture_timeslot(chromosome), decode_first_lecture_room(chromosome), population, decode_section(chromosome))
    conflicts += same_section_assigned_different_rooms_at_same_time(decode_second_lecture_day(chromosome), decode_second_lecture_timeslot(chromosome), decode_second_lecture_room(chromosome), population, decode_section(chromosome))

    #A room cannot be assigned for two different sections at the same time.
    conflicts += same_room_assigned_to_different_sections_at_same_time(decode_first_lecture_day(chromosome), decode_first_lecture_timeslot(chromosome), decode_first_lecture_room(chromosome), population, decode_section(chromosome))
    conflicts += same_room_assigned_to_different_sections_at_same_time(decode_second_lecture_day(chromosome), decode_second_lecture_timeslot(chromosome), decode_second_lecture_room(chromosome), population, decode_section(chromosome))

    #No Professor can teach more than 2 courses
    conflicts += professor_course_count(decode_professor(chromosome), professor_courses)

    #No Section can have more than 5 courses in a semester
    conflicts += section_course_count(decode_course(chromosome), decode_section(chromosome), section_courses)

    #Each course would have two lectures per week not on the same or adjacent days
    conflicts += course_have_two_lectures_per_week_not_on_same_or_adjacent_days(decode_course(chromosome), population) 

    #Lab lectures should be conducted in two consecutive slots.
    conflicts += lab_lectures_two_consective_time_slots(decode_course_type(chromosome), decode_first_lecture_timeslot(chromosome), decode_second_lecture_timeslot(chromosome), decode_first_lecture_day(chromosome), decode_second_lecture_day(chromosome), decode_first_lecture_room(chromosome), decode_second_lecture_room(chromosome))

    return conflicts
```

### Fitness Function

```python
def fitness_function(population):
    room_usage = {}
    professor_courses = {}
    section_courses = {}
    fitness_scores = []

    for chromosome in population:
       conflicts = calculate_conflicts(chromosome, population, room_usage, professor_courses, section_courses)
       fitness_scores.append(-1*conflicts)

    return fitness_scores
```

### Tournament Selection

```python
def tournament_selection(population, fitness_scores, k = 20):
    length = len(population) - 1
    selected_idex = random.randint(0, length)
    iter = k - 1
    for _ in range(iter):
        length = len(population)
        num = length - 1
        idex = random.randint(0, num)
        if fitness_scores[idex] > fitness_scores[selected_idex]:
            selected_idex = idex

    store = population[selected_idex]
    return store
```

### Crossover

```python
def crossover(parent1, parent2, r_cross):
    child1, child2 = copy.deepcopy(parent1), copy.deepcopy(parent2)
    len_courses = len(Courses)
    len_sections = len(Sections)
    len_professors = len(Professors)
    len_timeslots = len(Timeslots)
    len_rooms = len(Rooms)

    gene_sizes = [len_courses, 1, len_sections, 7, len_professors, 5, len_timeslots, len_rooms, 1, 5, len_timeslots, len_rooms]
    num_genes = gene_sizes.__len__()
    

    if random.random() < r_cross:
        chromosone_length = len(parent1)
        for i in range (1, num_genes + 1):
            sum_gene_sizes = sum(gene_sizes[:i]) 
        gene_end_indices = [sum_gene_sizes]
        gene_end_idx = random.choice(gene_end_indices)
        crossover_point = random.randint(gene_end_idx, min(chromosone_length - 1, gene_end_idx))
        p1 = parent1[:crossover_point]
        p2 = parent2[:crossover_point]
        child1 = p1 + p2
        child2 = p2 + p1
        
        
    return copy.deepcopy([child1, child2])
```

### Defining Generation Segment

```python
def define_gene_segments():
    course_gene , course_type_gene = len(Courses) , 1
    section_gene, section_strength_gene = len(Sections), 7
    professor_gene, room1_days_gene = len(Professors), 5
    room1_timeslots_gene, room1_room_gene = len(Timeslots), len(Rooms)
    room1_capacity_gene, room2_days_gene = 1, 5
    room2_timeslots_gene, room2_room_gene = len(Timeslots), len(Rooms)
    room2_capacity_gene = 1

    gene_segments = [
        course_gene, course_type_gene, section_gene, section_strength_gene,
        professor_gene, room1_days_gene, room1_timeslots_gene,
        room1_room_gene, room1_capacity_gene, room2_days_gene,
        room2_timeslots_gene, room2_room_gene, room2_capacity_gene
    ]

    return copy.deepcopy(gene_segments)
```

### Mutation

```python

def mutation(chromosome, mut_rate):
   
    mutated_chromosome = copy.deepcopy(chromosome)
    if random.random() >= mut_rate:
        return mutated_chromosome
    section_strength_gene = 7

    gene_segments = copy.deepcopy(define_gene_segments())

    length = len(gene_segments) - 1
    gene_index = random.randint(0, length)
    sum_gene_sizes = sum(gene_segments[:gene_index])
    segment_start = sum_gene_sizes
    segment_end = gene_segments[gene_index] + segment_start 

    if (gene_segments[gene_index] == section_strength_gene): 
        r = random.randint(segment_start, segment_end - 1)
        mutated_chromosome[r] = 1 if mutated_chromosome[r] == 0 else 0

    elif segment_end - segment_start == 1:
        mutated_chromosome[segment_start] = 1 if mutated_chromosome[segment_start] == 0 else 0
            
    else:
        try:
            sublist = mutated_chromosome[segment_start : segment_end]
            flip = random.randint(segment_start, segment_end - 1)
            mutated_chromosome[sublist.index(1) + segment_start] = 0
            mutated_chromosome[flip] = 1
        except ValueError:
            temp = 0
            print("ValueError", temp)
            pass

    return mutated_chromosome
```

### Creating Next Generation

```python
def check_odd(var):
    return var & 1

def create_next_generation(selected, r_cross, r_mut):
    children = []
    len_of_selcted = len(selected)
    num_parents = len_of_selcted

    if check_odd(num_parents):
        num_parents -= 1

    two = 2
    for i in range(0, num_parents, two):
        first_ind = i
        second_ind = i + 1
        p1 = selected[first_ind]
        p2 = selected[second_ind]
        for child in crossover(copy.deepcopy(p1), copy.deepcopy(p2), r_cross):
            children.append(copy.deepcopy(mutation(child, r_mut)))

    return children
```

### Genetic Algorithm

```python
def genetic_algorithm(objective, n_iter, n_pop, r_cross, r_mut, population):
    best = []
    best_from_each_generation = []
    best_eval = float('-inf')
    best_val = float('-inf')
    best_fitness_generation = 0
    var = (objective(population))
    for i in var:
        best_fitness_generation += i
    best_generation = copy.deepcopy(population)


    for gen in range(n_iter):
        fitness_scores = objective(population)
        print(gen)

        for i in range(n_pop):
            if fitness_scores[i] >= best_val:
                    best_val = fitness_scores[i]
                    best_from_each_generation.append(population[i])

            if fitness_scores[i] > best_eval:
                best, best_eval = population[i], fitness_scores[i]
                print("Generation No: %d" % (gen))
                print("New Best Fitness = %.3f" % (fitness_scores[i]))
                print("Best Chromosome: ")
                print(decode_chromosome(best))


        selected = [tournament_selection(population, fitness_scores) for _ in range(n_pop)]

        children = create_next_generation(selected, r_cross, r_mut)
        var = sum(objective(children))

        if (var > best_fitness_generation):
            best_generation = copy.deepcopy(children)
            best_fitness_generation = var

        population = copy.deepcopy(children)
        n_pop = len(population)

    winner = list(map(lambda x: x, best_generation))
    print(f"The best fitness population is {sum(objective(winner))}")
    print("The best generation after is : ")
    for i, c in enumerate(winner):
        print(i + 1, " decoded - > chromosome  is : ", decode_chromosome(c))
        print(i + 1, " encoded - > chromosme i s: ", c) 

    for c in best_from_each_generation:
        print("Best from each generation: ", decode_chromosome(c))
    return [best, best_eval]
```

```python

def main():
    
    population_size = 10
    n_iter = 100
    r_cross = 0.3
    r_mut = 0.9

    
    population = [generate_random_chromosomes() for _ in range(population_size)]


    print(f"Fitness of population before : {sum(fitness_function(population))}")
    print("Population Before :\n\n")
    for chro in population:
        print(decode_chromosome(chro))

    best_solution, best_fitness = genetic_algorithm(fitness_function, n_iter, len(population), r_cross, r_mut, population)

    print("\n##############################################\n")
    print("The best chromosome throughout the entire simulation is : ")
    print( decode_chromosome(best_solution))
    print("The fitness of the best chromosome throughout the entire simulation is : ", best_fitness)

if __name__ == "__main__":
    main()
```
